```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel("Employee Sample Data - A(1).xlsx")
print(df.head())
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.duplicated().sum())
```

```python
print(df.isnull().sum())
```

```python
df["Annual Salary"].fillna(df["Annual Salary"].mean(), inplace=True)
df["Age"].fillna(df["Age"].mean(),inplace=True)
df["Age"].astype(int)
df["Ethnicity"].fillna("Unknown", inplace=True)
df.dropna(subset=["Department", "Gender", "Country"], inplace=True)
df["Bonus %"].fillna(0, inplace=True)
```

```python
df.shape
```

```python
# number of employees in each department
plt.bar(df["Department"].value_counts().index, df["Department"].value_counts().values)
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees per Department")
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
# number of employees in each business unit
plt.bar(df["Business Unit"].value_counts().index, df["Business Unit"].value_counts().values)
plt.xlabel("Business Unit")
plt.ylabel("Number of Employees")
plt.title("Number of Employees per Business Unit")
plt.tight_layout()
plt.show()
```

```python
plt.hist(df["Age"], bins=10)
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Distribution of Employees by Age")
plt.tight_layout()
plt.show()
```

```python
plt.hist(df["Annual Salary"], bins=10)
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.title("Distribution of Employees by Salary")
plt.tight_layout()
plt.show()
#
```

```python
#Gender distribution
plt.pie(df["Gender"].value_counts(), labels=df["Gender"].value_counts().index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
#Gender distribution
plt.pie(df["Department"].value_counts(), labels=df["Department"].value_counts().index, autopct="%1.1f%%")
plt.title("Department Distribution")
plt.show()
```

```python
plt.bar(df["Country"].value_counts().index, df["Country"].value_counts().values)
plt.xlabel("Country")
plt.ylabel("Number of Employees")
plt.title("Number of Employees per Country")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x="Gender", y="Annual Salary", data=df)
plt.title("Annual Salary Distribution by Gender")
plt.show()
#
```

```python
sns.boxplot(x="Ethnicity", y="Bonus %", data=df)
plt.title("Bonus Distribution by Ethnicity")
plt.show()
```

```python
sns.boxplot(x="Gender", y="Bonus %", data=df)
plt.title("Bonus Distribution by Gender")
plt.show()
```

```python
plt.bar(df["Department"].value_counts().index, df.groupby("Department")["Annual Salary"].mean())
plt.xlabel("Department")
plt.ylabel("Average Annual Salary")
plt.title("Average Annual Salary per Department")
plt.tight_layout()
plt.xticks(rotation=45)
```

```python
plt.pie(df["Ethnicity"].value_counts(), labels=df["Ethnicity"].value_counts().index, autopct="%1.1f%%")
plt.title("Ethnicity Distribution")
plt.show()
#
```

```python
df["Status"]= df["Exit Date"].apply(lambda x: "Active" if pd.isnull(x) else "Inactive")
```

```python
plt.pie(df["Status"].value_counts(), labels=df["Status"].value_counts().index, autopct="%1.1f%%")
plt.title("Status Distribution")
plt.show()
```

```python
plt.bar(df["Status"].value_counts().index, df["Status"].value_counts().values)
plt.xlabel("Status")
plt.ylabel("Number of Employees")
plt.title("Number of Employees per Status")
plt.tight_layout()
plt.show()
#
```

```python
plt.bar(df["Business Unit"].value_counts().index, df.groupby("Business Unit")["Annual Salary"].mean())
plt.xlabel("Business Unit")
plt.ylabel("Average Annual Salary")
plt.title("Average Annual Salary per Business Unit")
plt.tight_layout()
```

```python
plt.bar(df["Status"].value_counts().index, df.groupby("Status")["Annual Salary"].mean())
plt.xlabel("Status")
plt.ylabel("Average Annual Salary")
plt.title("Average Annual Salary per Status")
plt.tight_layout()
#
```

```python
employees_with_bonus = df[df['Bonus %'] > 0]
bonus_counts_per_department = employees_with_bonus['Department'].value_counts()

plt.figure(figsize=(8, 4))
plt.bar(bonus_counts_per_department.index, bonus_counts_per_department.values)
plt.xlabel("Department")
plt.ylabel("Number of Employees with Bonus")
plt.title("Number of Employees with Bonus per Department")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
